## Test of relationship using Spearman r

After computing the mean and visualizing the data separately, we will now use Spearman correlation to see the correlation and if there is a significant relationship between technological skills and academic performance of senior high school students.

First, let us make a dataframe containing the general average

```python
genAve_df.index = range(1,51)
```

Then make another dataframe containing the mean of different skills 

```python
mean_df = pd.concat([response_df.iloc[0:,0:5].mean(axis=1),response_df.iloc[:,5:10].mean(axis=1),response_df.iloc[:,10:15].mean(axis=1),response_df.iloc[:,15:20].mean(axis=1)], axis=1)
```

Then, let us combine the two dataframe

```python
mean_ave_df = pd.concat([mean_df, genAve_df], axis=1) 
mean_ave_df.columns = ['Typing skills', 'Browser skills', 'Using office', 'Security and privacy', 'General average']
```

```python
result = mean_ave_df.corr(method='spearman').round(decimals=2)
```

```python
new_result = result.iloc[4,:4]
```

## Plotting of data

```python
labels = ['Typing', 'Browser', 'Office', 'Security']
values = new_result

plt.figure(figsize=(5,3))
corr_graph = plt.bar(labels,values)
plt.bar_label(corr_graph, new_result)
plt.title('Comparison of the Correlation between' "\n" 'technological skills and academic performance')
plt.xlabel("Technological skills")
plt.ylabel("Correlation")
plt.ylim(-0.04, 0.1);
```

<b> Note: </b> 
*Correlation: <br>
1.00 (Perfect Positive/Negative Relationship), <br>
0.91 to 0.99 (Very Strong Negative/Positive Relationship), <br>

0.41 to 0.70 (Moderate Positive/Negative Relationship), <br>
0.31 to 0.40 (Low Positive/Negative Relationship), <br>
0.01 to 0.30 (Very Low Positive/Negative Relationship), <br>
0.00 (No/Negligible Relationship)

## Interpretation

Typing skills have the highest correlation (0.095), followed by security and privacy (0.04), browsing skills (0.01), and finally office skills (-0.02). It also shows that the three skillstyping skills, browsing skills, and security and privacy skillshave a very low positive relationship except for office skills, which have a very low negative relationship. 
It also shows that every technological skill accepts the null hypothesis, which is "There is no significant relationship between the technological skills of the students and their academic performance."

## Conclusion

It can be concluded that even though technological skills can help students in various ways, such as allowing them to access information more easily and communicate more effectively, it does not necessarily have a significant effect on their academic performance. There are various factors that contribute to a student's academic performance, such as their motivation, study habits, and aptitude for the subject matter, and while technological skills may be beneficial, they are not the sole determinant of a student's academic success.
